```python
title = []
question = []
de = []
answer = []
for i in range(0,3):
    for j in range(1,11):
        driver.find_element_by_xpath('//*[@id="CP1_COM_Page_Controllor_spnPage%s"]' % j).click()
        time.sleep(1)
        for k in range(1,11):
            division = driver.find_element_by_xpath('//*[@id="CP1_iuwgList"]/tbody/tr[%s]/td[1]/span' % k).text
            if division == 'A':
                driver.find_element_by_xpath('//*[@id="CP1_iuwgList_jemog_%s"]' % (k-1)).click()
                time.sleep(1)
                page_title =  driver.find_element_by_xpath('//*[@id="CP1_lblTitle"]').text
                print(page_title)
                page_question = driver.find_element_by_xpath('//*[@id="CP1_lblQContent"]').text
                print(page_question)
                page_de = driver.find_element_by_xpath('//*[@id="CP1_lblAnswerBuseo"]').text
                print(page_de)
                page_answer = driver.find_element_by_xpath('//*[@id="CP1_lblAnswerContent"]').text
                print(page_answer)
                title.append(page_title)
                question.append(page_question)
                de.append(page_de)
                answer.append(page_answer)
                driver.find_element_by_xpath('//*[@id="CP1_btnList"]').click()
                time.sleep(1)
    driver.find_element_by_xpath('//*[@id="CP1_COM_Page_Controllor_lbtnNext10"]').click()
    time.sleep(1)
```
